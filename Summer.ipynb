# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("summer.csv")
```

### 1. In how many countries Summer Olympics is held so far?

```python
n=pd.unique(df['Country'])
print("Answer:",len(n))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
##Finding counts of gold with respect to the sports

d={}
for i in range(len(df["Medal"])):
    if df["Medal"][i]=="Gold":
        if df["Sport"][i] not in d: 
            d[df["Sport"][i]]=1
        else:
            d[df["Sport"][i]]+=1
```

```python
##Graph
sport = list(d.keys())
medals = list(d.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(sport, medals, color ='maroon',
        width = 0.4)
 
plt.xlabel("Sport")
plt.xticks(rotation=90)
plt.ylabel("Medal")
plt.title("Count of the GOLD medals")
plt.show()
```

### Top 5

```python
sport_order = sorted(d.items(), key=lambda x: x[1],reverse=True)
print(sport_order[:5])
```

### 4. Which player has won most number of medals? (Top 5)

```python
d3={}
for i in range(len(df["Athlete"])):
    if df.Medal[i]=="Gold" or df.Medal[i]=="Silver" or df.Medal[i]=="Bronze":
        if df.Athlete[i] not in d3:
            d3[df.Athlete[i]]=1
        else:
            d3[df.Athlete[i]]+=1
```

```python
sport_order = sorted(d3.items(), key=lambda x: x[1],reverse=True)
print(sport_order[:5])
```

```python
##Graph
athlete=[]
medals=[]
for i in range(0,5):
    athlete.append(sport_order[i][0])
    medals.append(sport_order[i][1])


fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(athlete, medals, color ='maroon',
        width = 0.4)
 
plt.xlabel("Athlete")
plt.xticks(rotation=90)
plt.ylabel("Medals")
plt.title("Popularity")
plt.show()
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
##Counter

d5={}
for i in range(len(df.Event)):
    if df.Event[i] not in d5:
        d5[df.Event[i]]=1
    else:
        d5[df.Event[i]]+=1
```

```python
#Top 5

sport_order = sorted(d5.items(), key=lambda x: x[1],reverse=True)
print(sport_order[:5])
```

```python
#Graph

sport=[]
count=[]
for i in range(0,5):
    sport.append(sport_order[i][0])
    count.append(sport_order[i][1])


fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(sport, count, color ='maroon',
        width = 0.4)
 
plt.xlabel("Sport")
plt.xticks(rotation=90)
plt.ylabel("Medal")
plt.title("Popularity")
plt.show()
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
#Calculate counts
d5={}
for i in range(len(df.Event)):
    if df.Sport[i] not in d5 and df.Gender[i]=="Women":
        d5[df.Sport[i]]=1
    elif df.Gender[i]=="Women":
        d5[df.Sport[i]]+=1
```

```python
#Find top 5
sport_order = sorted(d5.items(), key=lambda x: x[1],reverse=True)
print(sport_order[:5])
```

```python
#Graph

sport=[]
count=[]
for i in range(0,5):
    sport.append(sport_order[i][0])
    count.append(sport_order[i][1])

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(sport, count, color ='maroon',
        width = 0.4)
 
plt.xlabel("Sport")
plt.xticks(rotation=90)
plt.ylabel("Count")
plt.title("Number of women participants")
plt.show()
```
